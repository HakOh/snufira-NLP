## myngrams 함수 만들기

```python
def myngrams(sent, n, boundary=True):
    sent = sent.split()
    if boundary:
        sent = ['<s>'] + sent + ['</s>']
    ngram_list = [sent[i:i+n] for i in range(len(sent)-n+1)]
    
    return ngram_list
```

```python
sent = 'I almost got late, but luckily I arrived on time.'
# sent.split()
myngrams(sent, 3, True)
```

## stopwords 제거 후 N-gram 사용하기

http://www.nltk.org/book/ch02.html

There is also a corpus of stopwords, that is, high-frequency words like the, to and also that we sometimes want to filter out of a document before further processing. Stopwords usually have little lexical content, and their presence in a text fails to distinguish it from other texts.

```python
import nltk
from nltk.corpus import brown

news_text = brown.words(categories='news')
fdist = nltk.FreqDist(w.lower() for w in news_text) # FreqDist에 단어의 배열이 들어가면, 각 단어가 몇 번 등장했는지 알려준다.
```

```python
values = list(fdist.values())
values.sort()
values.reverse()
values
```

```python
fdist['a']
```

```python
fdist = nltk.FreqDist(ngrams(w.lower(), 2) for w in news_text)
```

```python
from nltk.corpus import stopwords
stopwords.words('english')

sent = 'There is also a corpus of stopwords, that is, high-frequency words like the, to and also that we sometimes want to filter out of a document before further processing.'
sent = sent.lower().split()
print('### stopword 제거 전')
print(list(nltk.bigrams(sent)))
sent = [item for item in sent if item not in stopwords.words('english')]
print('### stopword 제거 후')
print(list(nltk.bigrams(sent)))
```

Q.17 ◑ Write a function that finds the 50 most frequently occurring words of a text that are not stopwords.

```python
import nltk
from nltk.corpus import brown
from nltk.corpus import stopwords
import re
```

```python
sent = 'The book found yesterday is not my book, but hers'
sent = sent.lower().split()

sent =[w for w in sent if w not in stopwords.words('english')]
sent
```

```python
def fiftymost(words, n):
    '''
    words : token으로 구성된 리스트
    n : 전체 단어 중에서 n번째로 많이 등장한 횟수
    '''
    # list에 있는 단어를 소문자로 만든다.
    words = [w.lower() for w in words]
    # list에 있는 단어 중에 stopwords는 제외한다.
    words = [w for w in words if w not in stopwords.words('english')]
    
    # list에 있는 단어 중 비단어는 제외한다.
    pattern = r'^\W'
    words = [w for w in words if not re.search(pattern, w)]
    
    # FreqDist에 단어의 배열이 들어가면, 각 단어가 몇 번 등장했는지 알려준다.
    fdist = nltk.FreqDist(words) 
    
    # 등장횟수를 기준으로 내림차순으로 정렬한다.
    # sorted() 함수 설명
    # - 정렬 대상 : fdist.items()
    # - 정렬 기준 : (key, value) 튜플에서 1번째 index인 value를 기준으로 정렬하되, reverse=True 옵션 적용
    fiftymost = sorted(fdist.items(), key=lambda x: x[1], reverse=True)
    
    for i, w in enumerate(fiftymost[:n]):
        print("{}번째 : {}".format(i+1, w))
    
    # return fiftymost[:n]
        
```

```python
news_text = brown.words(categories='news')   
fiftymost(news_text, 50)
```

Q.18 ◑ Write a program to print the 50 most frequent bigrams (pairs of adjacent words) of a text, omitting bigrams that contain stopwords.

```python
def fiftymost_bigrams(words, n, boundary=True):
    '''
    words : token으로 구성된 리스트
    n : 전체 단어 중에서 n번째로 많이 등장한 횟수
    '''
    
    # list에 있는 단어를 소문자로 만든다.
    words = [w.lower() for w in words]
    # list에 있는 단어 중에 stopwords는 제외한다.
    words = [w for w in words if w not in stopwords.words('english')]
    
    # list에 있는 단어 중 비단어는 제외한다.
    pattern = r'^\W'
    words = [w for w in words if not re.search(pattern, w)]
    
    if boundary:
        words = ['<s>'] + words + ['</s>']
    bigram_list = [words[i:i+2] for i in range(len(words)-2+1)]
    
    ############# 여기서부터 잘 모르겠음
    cfd = nltk.ConditionalFreqDist(bigram_list) 
    
    # 50번째 많이 등장한 단어를 찾기 위해 전체 단어의 등장횟수를 내림차순으로 정리한다.
    values = list(cfd.values())
    values.sort()
    values.reverse()
    
    # print("{}번째로 많이 등장한 횟수: {}".format(n, values[n-1]))
    
    # 등장횟수를 기준으로 내림차순으로 정렬한다.
    # sorted() 함수 설명
    # - 정렬 대상 : fdist.items()
    # - 정렬 기준 : (key, value) 튜플에서 1번째 index인 value를 기준으로 정렬하되, reverse=True 옵션 적용
    fiftymost = sorted(cfd.items(), key=lambda x: x[1], reverse=True)
    
    for i, w in enumerate(fiftymost[:n]):
        print("{}번째 : {}".format(i+1, w))
```

```python
news_text = brown.words(categories='news')   
fiftymost_bigrams(news_text, 50, True)
```
